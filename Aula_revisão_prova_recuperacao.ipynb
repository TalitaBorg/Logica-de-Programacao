```python
A = [
    [1,2],
    [3,4]
]

B = [
    [5,6],
    [7,8]
]

SomaMatriz = []

#Somando elemento por elemento
for linhas in range(len(A)):
  linha = []
  for colunas in range(len(A[0])):
    soma = A[linhas][colunas] + B[linhas][colunas]
    linha.append(soma)
  SomaMatriz.append(linha)
print("Matriz soma: ")
for linha in SomaMatriz:
  print(linha)
```
