```python
A = [
    [1,2,3],
    [4,5,6]
]

B = [
    [4,5,6],
    [7,8,9]
]

Soma_matriz = []

#Somando elemento por elemento
for linhas in range(len(A)):
  linha = []
  for colunas in range(len(A[0])):
    soma = A[linhas][colunas] + B[linhas][colunas]
    linha.append(soma)
  Soma_matriz.append(linha)
print("Matriz soma: ")
for linha in Soma_matriz:
  print(linha)
```

```python
# Multiplicação de matriz 2x3 por 3x2

# Matriz A (2x3)
A = [
    [1, 2, 3],
    [4, 5, 6]
]

# Matriz B (3x2)
B = [
    [7, 8],
    [9, 10],
    [11, 12]
]

# Matriz resultado (2x2)
resultado = [[0, 0],
             [0, 0]]

# Multiplicação de matrizes
for i in range(2):        # linhas de A
    for j in range(2):    # colunas de B
        for k in range(3):  # elementos comuns
            resultado[i][j] += A[i][k] * B[k][j]

# Exibir resultado
print("Matriz Resultante (2x2):")
for linha in resultado:
    print(linha)
```

```python
#Subtração de matrizes de mesma ordem
A = [
    [4,8],
    [10,12]
]

B = [
    [1,4],
    [8,5]
]

Subrai_matriz = []

#Subtraindo elemento por elemento
for linhas in range(len(A)):
  linha = []
  for colunas in range(len(A[0])):
    subtrai = A[linhas][colunas] - B[linhas][colunas]
    linha.append(subtrai)
  Subrai_matriz.append(linha)
print("Matriz subtração: ")
for linha in Subrai_matriz:
  print(linha)
```

```python
linhas = int(input("Digite o número de linhas da matriz: "))
colunas = int(input("Digite o número de colunas da matriz: "))

#Lendo a primeira matriz
print("Digite os elementos da matriz A: ")
A = []
for i in range(linhas):
  linha = []
  for j in range(colunas):
    valor = int(input(f"A[{i}][{j}]: "))
    linha.append(valor)
  A.append(linha)

#Lendo a segunda matriz
print("Digite os elementos da matriz B: ")
B = []
for i in range(linhas):
  linha = []
  for j in range(colunas):
    valor = int(input(f"B[{i}][{j}]: "))
    linha.append(valor)
  B.append(linha)

print(f"As matrizes são A: {A} e B: {B}")
```
