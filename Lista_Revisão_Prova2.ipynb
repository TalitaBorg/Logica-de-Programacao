```python
#Questão 5

def maior_numero (numero1, numero2, numero3):
  if(numero1 > numero2):
    maior = numero1
  else:
    maior = numero2
  if(maior > numero3):
    maior = maior
  else:
    maior = numero3

  return maior

if __name__ == '__main__':
  m = maior_numero(6,8,2)
  print(m)

```

```python
#Verifica se o número é primo

def numero_primo(numero):
  if numero <= 1:
    return False # 0 e 1 não são números primos
  for i in range(2, numero):
    if numero % i == 0:
      return False # Encontrou um divisor, não é primo
  return True # Nenhum divisor encontrado, é primo

if __name__ == '__main__':
  m = numero_primo(2)
  print(m)
```

```python
def numero_inteiro(numero):
  if(numero % 2 ==0):
    return True
  else:
    return False

if __name__ == '__main__':
  numero = int(int(input("Digite um número: ")))
  m = numero_inteiro(numero)
  print(m)
```

```python
def funcao_area_circulo(raio):
  pi = 3.14
  area = (pi * (raio**2))
  return area

if __name__ == '__main__':
  m = funcao_area_circulo(4)
  print(m)
```
