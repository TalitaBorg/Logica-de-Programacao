```python
conta_par = 0
conta_impar = 0
contador = 0
soma_par = 0

while (contador < 10):
  numero = int(input("Digite um número: "))
  if(numero % 2 == 0):
    conta_par = conta_par + 1 #conta os números pares
    soma_par = soma_par + numero
  else:
    conta_impar = conta_impar + 1 #conta os números ímpares
  contador = contador + 1
#print(f"Foram digitados {conta_par} números pares") #imprime na tela números pares
#print(f"Foram digitados {conta_impar} números ímpares") #imprime na tela números ímpares
print(f"Foram digitados {conta_par} números pares e a soma deles é {soma_par}")
if (conta_par > 0):
  media = (soma_par)/conta_par
  print(f"A média dos números pares é {media}")
```

```python
conta_par = 0
conta_impar = 0
multiplica_par = 1
multiplica_impar = 1
contador = 0
quantidade = 5 #quantidade de nº que são lidos
while (contador < quantidade):
  numero = int(input("Digite um número: "))
  if (numero % 2 == 0):
    conta_par = conta_par + 1 #conta quantos nº pares o usuário digitou
    multiplica_par = multiplica_par * numero #multiplica valores pares
  else:
    conta_impar = conta_impar + 1 #conta quantos nº ímpares o usuário digitou
    multiplica_impar = multiplica_impar * numero #multiplica valores ímpares
  contador = contador + 1
  if (conta_par > 0):
    print(f"A multiplicação dos números pares é {multiplica_par}")
  else:
    print(f"Nenhum número par foi digitado!")
  if(conta_impar > 0):
    print(f"A multiplicação dos números ímpares é {multiplica_impar}")
  else:
    print(f"Nenhum número ímpar foi digitado!")
```
