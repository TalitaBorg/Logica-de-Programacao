```python
#Exercício 06

nota1 = float(input("Digite uma nota: "))
nota2 = float(input("Digite uma nota: "))
nota3 = float(input("Digite uma nota: "))

media = (nota1 + nota2 + nota3)/3

if(media>=6):
  print(f"Sua média é {media}, você está aprovado!")
else:
  print(f"Sua média é {media}, abaixo de 6, logo você está reprovado!")
```

```python
x = 1
while (x <= 4):
  numero = int(input("Digite um número: "))
  if (numero % 2 == 0): #testa se o numero é par
    quadrado_numero = numero ** 2
    print(quadrado_numero)
  else: #testa se o numero é impar
    raiz_quadrada = numero ** 0.5
    print(raiz_quadrada)
  x = x + 1



```

```python
x = 0 #inicializa variavel x com zero
conta_par = 0 # inicializa contador

while(x < 10): #condição de parada são 10 números para pedir ao usuário
  numero = int(input("Digite um número: "))
  if (numero % 2 == 0): #testa se o número é par
    conta_par = conta_par + 1 #conta quantos números pares o usuário digitou
  x = x + 1  #incrementar o x
print(f"Foram digitados {conta_par} números pares")
```

```python
x = 0
conta_impar = 0

while(x < 10):
  numero = int(input("Digite um número: "))
  if (numero % 2 != 0):
    conta_impar = conta_impar + 1
  x = x + 1
print(f"Você digitou {conta_impar} números ímpares")
```

```python
numero = int(input("Digite um número da Tabuada: "))
contador = 0

while(contador < 10):
  Tabuada = numero + contador
  contador = contador + 1
  print(f"{numero} + {contador} = {Tabuada}")
```
